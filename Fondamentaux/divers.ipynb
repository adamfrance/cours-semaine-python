```python
# break
x = 100
while x >= 100:
    x += 1
    if x % 19 == 0:
        print(x, 'est le 1ier nbe plus grand que 100 et divisible par 19')
        break
```

```python
# trouver le 1ier nombre divisible par 18 et 56
# entre 18 et 56
i = 1
diviseur1 = 18
diviseur2 = 56
while True:
    if i % diviseur1 == 0 and i % diviseur2 == 0:
        print(f'Le plus petit multiple commun de {diviseur1} et de {diviseur2} est {i}')
        break
    i += 1
```
